```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

x = sp.Symbol('x', real=True)
y = sp.Symbol('y', real=True)
z = sp.Symbol('z', complex=True)

z = x + sp.I*y

U = 1
Q = -2

w = U*z + Q / 2.0 / sp.pi * sp.log(z)

psi = sp.im(w)

psi = sp.lambdify((x, y), psi, 'numpy')

xx = np.linspace(-2.0, 2.0, 1000)
yy = np.linspace(-2.0, 2.0, 1000)
xx, yy = np.meshgrid(xx, yy)

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)
ax.set_aspect('equal')
ax.axis('off')

levs = np.linspace(-2, 2, 21)
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'

plt.contour(xx, yy, psi(xx, yy), colors='black', levels = levs)
plt.plot(0,0, marker="o", color="black")
```
